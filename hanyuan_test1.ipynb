```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
feature_matrix = pd.read_csv("https://raw.githubusercontent.com/Jae1654/ML_Project/main/jobs_feature.csv")
feature_matrix
```

```python
X = feature_matrix.drop(['fraudulent', 'company_profile_length', 'description_length', 'requirements_length', 'benefits_length'], axis=1)
X
```

```python
# try k-means to somehow reduce the number of features

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state=0).fit(X)
```

```python
telecom = np.zeros((10,))

for i in range(5362):
    cluster = kmeans.labels_[i]
    telecom[cluster] = telecom[cluster] + X['telecommuting'][i]

telecom
```
